# IDENTIFYING TRENDS

```python
df.columns
```

```python
df_grouped = df.groupby('date')[['total_cases', 'total_deaths', 'people_fully_vaccinated']].sum().reset_index()

plt.figure(figsize=(10,6))

plt.plot(df_grouped['date'], df_grouped['total_cases'], label='Cases', marker='o', linestyle='-', color='blue')
plt.plot(df_grouped['date'], df_grouped['total_deaths'], label='Deaths', marker='s', linestyle='--', color='red')
plt.plot(df_grouped['date'], df_grouped['people_fully_vaccinated'], label='Vaccinated', marker='^', linestyle='-', color='green')

plt.title('Trend Over Time For Cases, Deaths And Vaccinated')
plt.xlabel('Date')
plt.ylabel('Values')
plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='total_cases', y='positive_rate')
plt.title('Cases vs Positive rate')
plt.xlabel('Cases')
plt.ylabel('Positive Rate')
plt.grid(True)
plt.show()
```

```python
df_grouped = df.groupby('continent')[['total_deaths', 'total_cases', 'people_fully_vaccinated']].sum().reset_index()
print(df_grouped)
```

```python
df_grouped_melted = df_grouped.melt(id_vars='continent',
                                    value_vars=['total_deaths', 'total_cases', 'people_fully_vaccinated'],
                                    var_name='Metric',
                                    value_name='Value')
plt.figure(figsize=(10,6))
sns.barplot(data=df_grouped_melted, x='continent', y='Value', hue='Metric', palette='Set2')

plt.title('Comparison Of Cases, Deaths and Vaccinated By Continent')
plt.xlabel('Continent')
plt.ylabel('Values')
plt.legend(title='Metric')
plt.show()
```

```python
#df_grouped = df.groupby(['continent', 'location'])['total_cases'].sum().reset_index()

#top_location = df_grouped.groupby('continent').apply(lambda x: x.nlargest(10, 'total_cases')).reset_index(drop=True)
heatmap_data = df.pivot_table(index='continent', columns='location', values='total_cases', aggfunc='sum').fillna(0)

plt.figure(figsize=(16,8))
sns.heatmap(heatmap_data, cmap='Reds', linewidths=0.5)

plt.title('COVID-19 Cases by Continent and Location')
plt.xlabel('Location')
plt.ylabel('Continent')
plt.show()
```

```python
columns_to_aggregate = ['total_deaths', 'total_cases', 'people_fully_vaccinated', 'positive_rate']
continent_data = df.groupby('continent')[columns_to_aggregate].sum().reset_index()
print(continent_data)
```

```python

fig, axes = plt.subplots(2,2, figsize=(14,12))

columns = ['total_deaths', 'total_cases', 'people_fully_vaccinated', 'positive_rate']
titles = ['Deaths', 'Cases', 'Vaccinated', 'Positive Rate']

for i, col in enumerate(columns):
    ax = axes[i // 2, i % 2]
    ax.pie(
        continent_data[col],
        labels=continent_data['continent'],
        autopct='%1.1f%%',
        startangle=90,
        colors=plt.cm.Paired.colors
    )
    ax.set_title(titles[i])

plt.tight_layout()
plt.show()
    
```
